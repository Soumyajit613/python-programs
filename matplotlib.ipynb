1. Basic Line Plot

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]
plt.plot(x, y, marker='o')
plt.title('Basic Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
```

2. Bar Chart

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]
plt.bar(categories, values, color='skyblue')
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

3. Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)
plt.hist(data, bins=30, alpha=0.7, color='purple')
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

4. Scatter Plot

```python
import matplotlib.pyplot as plt

x = [5, 7, 8, 9, 10]
y = [10, 14, 15, 20, 25]
plt.scatter(x, y, color='red')
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
```

5. Pie Chart

```python
import matplotlib.pyplot as plt

sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart Example')
plt.axis('equal')
plt.show()
```

6. Stacked Bar Chart

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C']
men_means = [20, 35, 30]
women_means = [25, 32, 34]
plt.bar(labels, men_means, color='lightblue', label='Men')
plt.bar(labels, women_means, bottom=men_means, color='salmon', label='Women')
plt.title('Stacked Bar Chart')
plt.ylabel('Scores')
plt.legend()
plt.show()
```

7. Area Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 11)
y = np.random.randint(1, 10, size=10)
plt.fill_between(x, y, color='skyblue', alpha=0.5)
plt.title('Area Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

9. Heatmap

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.rand(10, 10)
plt.imshow(data, cmap='hot', interpolation='nearest')
plt.title('Heatmap Example')
plt.colorbar()
plt.show()
```

10. Error Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 10)
y = np.sin(x)
error = np.random.rand(10) * 0.1
plt.errorbar(x, y, yerr=error, fmt='o', capsize=5)
plt.title('Error Bar Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

11. Polar Plot

```python
import matplotlib.pyplot as plt
import numpy as np

theta = np.linspace(0, 2 * np.pi, 100)
r = np.abs(np.sin(2 * theta))
plt.polar(theta, r)
plt.title('Polar Plot Example')
plt.show()
```

12. 3D Scatter Plot

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)
ax.scatter(x, y, z)
ax.set_title('3D Scatter Plot Example')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```

13. Multiple Subplots

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(2)
axs[0].plot(x, y1, color='blue')
axs[0].set_title('Sine Function')
axs[1].plot(x, y2, color='red')
axs[1].set_title('Cosine Function')
plt.tight_layout()
plt.show()
```

14. Annotation Example

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y)
plt.title('Annotation Example')
plt.annotate('Max Value', xy=(5, 11), xytext=(4, 10),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.show()
```

15. Violin Plot

```python
import matplotlib.pyplot as plt
import numpy as np

data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.violinplot(data)
plt.title('Violin Plot Example')
plt.xticks([1, 2, 3], ['Group 1', 'Group 2', 'Group 3'])
plt.ylabel('Value')
plt.show()
```

16. Histogram with KDE

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = np.random.randn(1000)
sns.histplot(data, kde=True, color='blue', bins=30)
plt.title('Histogram with KDE Example')
plt.show()
```

17. Step Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 11)
y = np.random.randint(1, 10, size=10)
plt.step(x, y, where='mid', color='orange')
plt.title('Step Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

18. Quiver Plot

```python
import matplotlib.pyplot as plt
import numpy as np

Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2
plt.quiver(X, Y, U, V)
plt.title('Quiver Plot Example')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.show()
```

19. Logarithmic Scale Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1, 100, 100)
y = np.log(x)
plt.plot(x, y)
plt.xscale('log')
plt.title('Logarithmic Scale Example')
plt.xlabel('Log Scale X-axis')
plt.ylabel('Y-axis')
plt.show()
```
